```python
positions_table = []
for mapping_dict in POSITIONS:
    for keyword, result in mapping_dict.mapping_dict.items():
        positions_table.append(
            [
                keyword,
                result,
                mapping_dict.case_sensitive,
                mapping_dict.spaces_sensitive,
            ]
        )

pd.DataFrame(
    positions_table, columns=["keyword", "result", "case_sensitive", "spaces_sensitive"]
).set_index("keyword")
```
